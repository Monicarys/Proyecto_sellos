```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
import random
import pickle
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
```

```python
DATADIR = 'Sellos/' ##En este folder tengo los folders de mis imgenes
Categories = ['Dado', 'Dardo', 'Foco', 'Maceta', 'Oso', 'Pulpo', 'Rueda']##Estos son los nombres de cada folder con imgenes
for category in Categories:
    path = os.path.join(DATADIR, category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)##Obtener matrices de las imgenes
        img_Vec = (np.asarray(img_array)).flatten()##Convertir en vectores las matrices
        plt.imshow(img_array, cmap='gray')
        plt.show()
        break
    break
```

```python
img_array.shape
```

```python
IMG_SIZE=140
new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))##Hago que todas las imgenes tengan el mismo tamao 140x140
plt.imshow(new_array,cmap='gray')
```

```python
training_data = []

def create_training_data():
    for category in Categories:
        path = os.path.join(DATADIR, category)
        class_num = Categories.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                img_N_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                img_Vec = (np.asarray(img_N_array)).flatten()
                training_data.append([img_N_array,img_Vec, class_num])
            except Exception as e:
                pass
##sea crea el dataset con tres partes en cada entrada,
##[la matriz para poder imprimir la imgen, el vector con el que quiero usar KNN, la etiqueta que me dice a cual de las imgenes pertenece]            
create_training_data()
```

```python
random.shuffle(training_data)
```

```python
#training_data
```

```python
Z=[]##Para guardar las matrices
X=[]##Para guardar los vectores
y=[]##Para guardar las etiquetas


for image, features, label in training_data:
    Z.append(image) 
    X.append(features)
    y.append(label)
```

```python
#Guardo todo para no tener que hacer lo anterior otra vez
pickle_out = open('X.pickle','wb')
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open('y.pickle','wb')
pickle.dump(y, pickle_out)
pickle_out.close()

pickle_out = open('Z.pickle','wb')
pickle.dump(Z, pickle_out)
pickle_out.close()
```

```python
pickle_in = open('X.pickle','rb')
X = pickle.load(pickle_in)
pickle_in = open('y.pickle','rb')
y = pickle.load(pickle_in)
pickle_in = open('Z.pickle','rb')
Z = pickle.load(pickle_in)
```

```python
# Divido entre datos de entrenamiento y de prueba
X_train, X_test, y_train, y_test, Z_train, Z_test = train_test_split(X, y, Z, stratify=y, random_state=0)
```

```python
#X
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.decomposition import NMF
# Decimos cuantas componentes queremos y damos la semilla aleatoria
nmf = NMF(n_components=10, random_state=0)
# Decimos con respecto a que conjunto hacer el ajuste
nmf.fit(X_train)
# Transformamos los datos de entrenamiento y prueba
X_train_nmf = nmf.transform(X_train)
X_test_nmf = nmf.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_nmf, y_train)
knn.score(X_test_nmf, y_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_scaled, y_train)
knn.score(X_test_scaled, y_test)
```

```python
from sklearn.svm import SVC
svm = SVC(C=100)
# Entrenamos al modelo con los datos de entrenamiento escalados
svm.fit(X_train_scaled, y_train)
# Imprimimos el porcentaje de acierto aplicar el mtodo al conjunto de prueba escalado
svm.score(X_test_scaled, y_test)
```
